```python
import numpy as np
import pandas as pd
import csv
import pylab as pl
import seaborn as sns
import mpl_toolkits.basemap as bm
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

## Загружаем данные

```python
rooms_data = pd.read_csv('rooms.csv')
rooms_data.head()
```

## Документирование

### type_of_accommodation:

```python
one = OneHotEncoder(sparse = False)
le = LabelEncoder()
```

```python
data = rooms_data['type_of_accommodation']
a = one.fit_transform(le.fit_transform(data).reshape(-1, 1))
```

```python
print(dict(enumerate(le.classes_)))
```

```python
type_of_accommodation = pd.DataFrame(a, columns=['type_of_accommodation0', 'type_of_accommodation1'])
type_of_accommodation.head()
```

```python
rooms_data = rooms_data.join(type_of_accommodation)
rooms_data = rooms_data.drop(['type_of_accommodation'], 1)
rooms_data.head()
```
